```python
# 1사분위수, 3사분위수 구하기
# df.quantile()를 이용하면 분위수를 구할 수 있습니다. 하위 25%에 해당하는 1사분위수와 75%에 해당하는 3사분위수를 구하겠습니다.
```

```python
pct25 = mpg['hwy'].quantile(.25)
pct25
```

```python
pct75 = mpg['hwy'].quantile(.75)
pct75
```

```python
# 1사분위수와 3사분위수의 거리를 나타낸 IQR(inter quartile range, 사분위 범위)을 구합니다
iqr = pct75 - pct25
iqr
```

```python
# 하한, 상한 구하기
# 극단치의 경계가 되는 하한과 상한을 구합니다.
    # 하한 : 1사분위수보다 'IQR의 1.5배'만큼 더 작은 값
    # 상한 : 3사분위수보다 'IQR의 1.5배'만큼 더 큰 값
```

```python
pct25 - 1.5 * iqr # 하한
```

```python
pct75 + 1.5 * iqr # 상한
```
